## Data Exploration and Visualisation 

```python
import seaborn as sns

sns.countplot(x='clarity', data=df, order = ['I1', 'SI2' , 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'])
plt.title("Figure 1: Count for clarity of diamonds", fontsize=16)
```

Figure 1 is sorted from lowest quality clarity to highest quality it shows that the low middle clarity diamonds are most common with the count tapering off to the higher quality diamonds.

```python
sns.countplot(x='cut', data=df, order = ['Fair', 'Good', 'Very Good', 'Premium, Ideal'])
plt.title("Figure 2: Count for cut of diamonds", fontsize=16)
```

Premium cut is the most common diamond in the dataset.

```python
sns.distplot(df['price']).set_title('Figure 3: Histogram of Price', fontsize = 15)
```

The diamond prices are a unimodal right skewed distribution with the most common price being below 5000 US dollars.

```python
import seaborn as sns

plt.figure(figsize = (15, 10))
plt.scatter(df['carat'], df['price'])
plt.title('Figure 4: Scatterplot of Price and carat', fontsize = 15)
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show();
```

There is an obvious positive correlation between carat and price.

```python
plt.figure(figsize = (15, 10))
plt.scatter(df['depth'], df['price'])
plt.title('Figure 5: Scatterplot of Price and Depth', fontsize = 15)
plt.xlabel("Depth")
plt.ylabel("Price")
plt.show();
```

There is no correlation between depth and price.

```python
plt.figure(figsize = (15, 10))
plt.scatter(df['table'], df['price'])
plt.title('Figure 7: Scatterplot of Price and Table', fontsize = 15)
plt.xlabel("Table")
plt.ylabel("Price")
plt.show();
```

There is no correlation between table and price

```python
plt.figure(figsize = (20, 10))
sns.boxplot(df['cut'], df['price'], hue = df['color'], order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], hue_order=['D', 'E', 'F', 'G', 'H', 'I', 'J'])
plt.title('Figure 7: Boxplot of Price broken down by Cut and Color', fontsize = 20)
plt.legend(loc='upper right')
```

The price increases from the left to the right in terms of cut with the exception of ideal which appears to drop in price a small amount compared to premium. The color is presented with D or E being the cheapest and J generally being the most expensive.

```python
plt.figure(figsize = (20, 10 ))
sns.boxplot(df['cut'], df['price'], hue = df['clarity'], order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], hue_order=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF' ])
plt.title('Figure 8: Boxplot of Price broken down by Cut and Clarity', fontsize = 20)
plt.legend(loc='upper right')
```

There appears to be no sensible order to the clarity of the diamonds 

```python
plt.figure(figsize= (20,10))
sns.scatterplot(df['carat'], df['price'], hue = df['color'])
plt.title('Figure 8: Scatterplot of Price by Distance coloured by House Type', fontsize = 15);
plt.legend(loc = 'upper right')
plt.show();
```

There is a far higher proportion of blue, purple and pink correlating to E, F and D color diamond towards the left side, where as the right side is mostly made up of orange, green and red correlating to I, J and H.
